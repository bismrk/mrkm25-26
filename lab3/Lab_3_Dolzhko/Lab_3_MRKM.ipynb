```python
from Crypto.PublicKey import ECC
from Crypto.Signature import DSS
from Crypto.Hash import SHA256

# 1. Генерація ключів ECDSA (крива NIST P-256)
key = ECC.generate(curve='P-256')

# Експорт ключів (для демонстрації)
private_key_pem = key.export_key(format='PEM')
public_key_pem = key.public_key().export_key(format='PEM')
print("Приватний ключ (PEM):\n", private_key_pem)
print("Публічний ключ (PEM):\n", public_key_pem)

# 2. Підпис повідомлення
message = b'Test message for ECDSA laboratory'
hash_obj = SHA256.new(message)

# Створення підписувача
signer = DSS.new(key, 'fips-186-3')  # Режим FIPS 186-3 (детермінований)
signature = signer.sign(hash_obj)
print("Підпис (hex):", signature.hex())

# 3. Верифікація підпису
# Новий хеш, бо sign споживає об'єкт
hash_obj = SHA256.new(message)

# Створення верифікатора (з публічним ключем)
verifier = DSS.new(key.public_key(), 'fips-186-3')
try:
    verifier.verify(hash_obj, signature)
    print("Підпис валідний!")
except ValueError:
    print("Підпис невалідний!")
```
